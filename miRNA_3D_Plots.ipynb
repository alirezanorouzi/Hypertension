```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import pandas as pd

path='/Users/alireza/OneDrive - University of Dundee/AN_ENSAT/Code/CleanedData2.CSV'

x=pd.read_csv(path)

label=x['Condition']
y=x['ConditionID']
x=x.drop(columns=['Unnamed: 0'])

#label=x['ConditionID']
x=x.drop(columns=['Condition'])
x=x.drop(columns=['ConditionID'])
```

```python
delcols='O2'
x=x.drop(columns=x.columns[x.columns.str[0:2]==delcols])

delcols='O4'
x=x.drop(columns=x.columns[x.columns.str[0:2]==delcols])


lda = LDA(n_components=4)
X_lda = lda.fit_transform(x, y)
```

```python
X_df=pd.DataFrame(X_lda)
X_df['label']=label
X_df.columns=['LDA1', 'LDA2','LDA3','LDA4','label']
```

```python
import plotly.express as px
fig = px.scatter_3d(X_df, x='LDA1', y='LDA2', z='LDA3',color='label')
fig.show()

```
